# show the lenght of ds.target

```python
import numpy as np
# using numpy to add the class to the a new dataset called data 
data = np.c_[ds.data,ds.target]
```

```python
col=ds.feature_names.copy()
col.append('class')
# adding the name class to the feature names
```

```python
col
```

# now load the new dataset data containing also the target into a pandas dataframe

```python
import pandas as pd
df=pd.DataFrame(#insert data here,columns=col)
```

# Now using data we'll try to plot some scatterplot using Matplotlib (the ones we see above), let's start from petal length vs petal width to see if the shape is the same.

```python
data[:,4]==0
#remember boolean selection vector, this can be used to filter only instances pertaining the specific class you want to plot
```

```python
# prep X and y for class 0 for petal length and petal width
X=data[data[:,4]==0][:,2]
y=data[data[:,4]==0][:,3]
# prep X and y for class 1 for petal length and petal width
X2=data[data[:,4]==1][:,2]
y2=data[data[:,4]==1][:,3]
# prep X and y for class 2 for petal length and petal width
X3=data[data[:,4]==2][:,2]
y3=data[data[:,4]==2][:,3]
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y,'ro',X2,y2,'bo',X3,y3,'go')
plt.show()
```
